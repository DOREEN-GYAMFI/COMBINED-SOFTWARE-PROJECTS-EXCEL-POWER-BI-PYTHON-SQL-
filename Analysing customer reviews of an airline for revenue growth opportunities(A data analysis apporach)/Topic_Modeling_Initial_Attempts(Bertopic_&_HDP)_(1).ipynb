```python
from bertopic import BERTopic
import pandas as pd
from bertopic import BERTopic
from gensim.models.coherencemodel import CoherenceModel
import pandas as pd
import nltk
from gensim.corpora import Dictionary
from gensim import corpora
```

```python
df = pd.read_csv('/content/all_reviews_with_sentiments(2018-2022).csv')
```

```python
df_2018 = df[df['Year']== 2018]
df_2018
```

```python
# embedding is changing a text into a numerical representation to improve it clustering
topic_model = BERTopic(embedding_model='all-MiniLM-L6-v2')
```

```python
#  'Review' is the name of the column containing the text data
text_data = df_2018['Review'].tolist()
```

```python
#
topics,probs =topic_model.fit_transform(text_data)
from bertopic.representation import KeyBERTInspired
```

```python
## Get the information from the model
# Topic -1 is outlier
topic_model.get_topic_info()
```

```python
# Get the frequency of each topic
topic_freq = topic_model.get_topic_freq()

# Print the number of topics
num_topics = len(topic_freq)
print("Number of topics for 2018:", num_topics)
```

```python
#get document that represent a topic
topic_model.get_representative_docs(0)
```

```python
# make a dataframe with every document and their corresponding topics
df1 = pd.DataFrame({"topic":topics,"document":text_data})
df1
```

```python
#Visualise the topics
# the closer they are, the more semantically similar they are
topic_model.visualize_topics()
```

```python
documents = pd.DataFrame({"Document": text_data,
                          "ID": range(len(text_data)),
                          "Topic": topics})
documents_per_topic = documents.groupby(['Topic'], as_index=False).agg({'Document': ' '.join})
cleaned_docs = topic_model._preprocess_text(documents_per_topic.Document.values)

# Extract vectorizer and analyzer from BERTopic
vectorizer = topic_model.vectorizer_model
analyzer = vectorizer.build_analyzer()

# Extract features for Topic Coherence evaluation
words = vectorizer.get_feature_names_out()
tokens = [analyzer(doc) for doc in cleaned_docs]
dictionary = corpora.Dictionary(tokens)
corpus = [dictionary.doc2bow(token) for token in tokens]
topic_words = [[words for words, _ in topic_model.get_topic(topic)]
               for topic in range(len(set(topics))-1)]

# Evaluate
coherence_model = CoherenceModel(topics=topic_words,
                                 texts=tokens,
                                 corpus=corpus,
                                 dictionary=dictionary,
                                 coherence='c_v')
coherence = coherence_model.get_coherence()
```

HDP FOR TOPIC MODELING

```python
import pandas as pd
from gensim.models import HdpModel
from gensim.corpora import Dictionary
from gensim.models import CoherenceModel
```

```python
# Load your data into a DataFrame
df = pd.read_csv('/content/all_reviews_with_sentiments(2018-2022).csv')
```

```python
# Filter out rows with the year 2018
df = df[df['Year'] != 2018]


# Tokenization
df['tokens'] = df['Review'].apply(lambda x: x.lower().split())

# Create a dictionary representation of the documents
dictionary = Dictionary(df['tokens'])

# Filter out words that appear in less than 2 documents or more than 80% of the documents
dictionary.filter_extremes(no_below=2, no_above=0.8)

# Convert the documents to bag-of-words (BoW) representation
corpus = [dictionary.doc2bow(tokens) for tokens in df['tokens']]

# Perform HDP topic modeling
hdp_model = HdpModel(corpus, id2word=dictionary)

# Print the topics
for idx, topic in hdp_model.print_topics():
    print(f"Topic {idx}: {topic}")
```

```python
# Get the total number of topics
num_topics = len(hdp_model.get_topics())
num_topics
```

```python
# Compute the coherence score
coherence_model_hdp = CoherenceModel(model=hdp_model, texts=df['tokens'], dictionary=dictionary, coherence='c_v')
coherence_hdp = coherence_model_hdp.get_coherence()
print(f"\nCoherence Score: {coherence_hdp}")
```
