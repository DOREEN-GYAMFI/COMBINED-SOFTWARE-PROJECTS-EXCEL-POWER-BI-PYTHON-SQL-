# 2018

## Distribution
checking the if topics in the year are normally distributed or not

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the merged dataset
merged_df = pd.read_csv('merged_data_after_topic_sentiment.csv')

# Convert sentiment_scores to numeric values
merged_df['sentiment_scores'] = pd.to_numeric(merged_df['sentiment_scores'])

# Get unique topic numbers for the year 2018
unique_topic_numbers_2018 = merged_df[merged_df['Year'] == 2018]['Topic'].unique()

# Loop through each topic number, creating separate histograms for 2018
for topic_number in unique_topic_numbers_2018:
    topic_data = merged_df[(merged_df['Year'] == 2018) & (merged_df['Topic'] == topic_number)]

    plt.figure(figsize=(4, 4))  # Adjust the figsize for smaller graphs

    # Create a histogram of sentiment scores for the current topic number in 2018
    plt.hist(topic_data['sentiment_scores'], bins=20, edgecolor='black')
    plt.xlabel('Sentiment Scores')
    plt.ylabel('Frequency')
    plt.title(f'Distribution of Sentiment Scores for Topic Number: {topic_number} - Year: 2018')
    plt.grid(True)

    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import numpy as np

# Create an empty DataFrame to store results
results_df = pd.DataFrame(columns=['Topic', 'Average_Sentiment', 'Permutation_Test_Statistic', 'P_Value', 'Hypothesis'])

# Filter data for the year 2018
year_2018_data = merged_df[merged_df['Year'] == 2018]

# Loop through each unique topic in the 'Topic' column for the year 2018
unique_topics_2018 = year_2018_data['Topic'].unique()
for topic in unique_topics_2018:
    # Extract sentiment scores for the current topic in 2018
    topic_sentiment_scores = year_2018_data.loc[year_2018_data['Topic'] == topic, 'sentiment_scores']

    # Extract the topic sentiment scores for the current topic
    topic_sentiment_topic = year_2018_data.loc[year_2018_data['Topic'] == topic, 'Topic_Sentiment_Score'].iloc[0]

    # Calculate the observed difference between average sentiment and topic sentiment
    observed_difference = np.mean(topic_sentiment_scores) - topic_sentiment_topic

    # Initialize variables to store permutation test results
    num_permutations = 10000  # Adjust as needed
    permutation_differences = np.empty(num_permutations)

    # Perform permutation test
    for i in range(num_permutations):
        shuffled_sentiments = np.random.permutation(topic_sentiment_scores)
        permutation_difference = np.mean(shuffled_sentiments) - topic_sentiment_topic
        permutation_differences[i] = permutation_difference

    # Calculate p-value
    p_value = np.mean(np.abs(permutation_differences) >= np.abs(observed_difference))

    alpha = 0.05

    # Determine whether to reject or accept the null hypothesis
    if p_value < alpha:
        hypothesis = "Reject"
    else:
        hypothesis = "Accept"

    # Append results to the results DataFrame
    results_df = results_df.append({'Topic': topic,
                                    'Average_Sentiment': topic_sentiment_topic,
                                    'Permutation_Test_Statistic': observed_difference,
                                    'P_Value': round(p_value, 3),  # Round p-value to 3 decimal places
                                    'Hypothesis': hypothesis}, ignore_index=True)

# Print the results DataFrame
results_df
```
