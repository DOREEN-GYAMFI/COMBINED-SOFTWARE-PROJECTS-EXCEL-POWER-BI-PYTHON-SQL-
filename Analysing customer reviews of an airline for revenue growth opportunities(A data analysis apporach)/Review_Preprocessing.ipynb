```python
#Expansion of contractions
def expand_contractions(text):
    words = []
    for word in text.split():
            words.append(contractions.fix(word))
    return ' '.join(words)

#Apply the expand_contractions function to the 'Review' column of the DataFrame
df['Review'] = df['Review'].apply(expand_contractions)
df
```

# **Text Substitution**

so the marker like symbols can be removed in this stage bold text

```python
# Text Substitution  so the marker like symbols can be removed in this stage
import re
def remove_markers(text):
    pattern = r"(@[A-Za-z0-9_]+)|[^\w\s]|#|http\S+"
    text = re.sub(pattern, '', text)
    return text

# apply the remove_markers function to the 'my_column' column
df['Review'] = df['Review'].apply(remove_markers)
df
```

```python
# remove numbers
def remove_numbers(text):
  pattern = r'\d+'
  text = re.sub(pattern,"",text)
  return text
```

```python
# apply the function to the 'my_column' column
df['Review'] = df['Review'].apply(remove_numbers)
df
```

```python
# Non-ACII removal
def replace_ascii(text):
  pattern = r'[^\x00-\x7F]+'
  text = re.sub(pattern,"",text)
  return text
```

```python
# apply the remove_markers function to the 'my_column' column
df['Review'] = df['Review'].apply(replace_ascii)
df
```

# **Tokenisation and Lemmatization with SpaCy**

```python
import pandas as pd
import spacy

# Load the English language model
nlp = spacy.load('en_core_web_sm')

# Function to lemmatize a sentence, convert past tense verbs to present tense, convert plural nouns to singular nouns,
# and convert adjectives and adverbs to their base forms
def lemmatize_and_convert(text):
    doc = nlp(text)
    converted_tokens = []
    for token in doc:
        # Lemmatize verbs, nouns, adjectives, and adverbs
        if token.pos_ in ('VERB', 'NOUN', 'ADJ', 'ADV'):
            converted_tokens.append(token.lemma_)
        else:
            converted_tokens.append(token.text)
    return ' '.join(converted_tokens)

if __name__ == "__main__":
    # Assuming you have loaded the DataFrame 'df' with a 'Review' column containing the sentences.

    # Apply the lemmatize_and_convert function to the 'Review' column
    df['Review'] = df['Review'].apply(lambda x: lemmatize_and_convert(str(x)))
```

# **Stop words removal with spacy**

```python
import spacy
import pandas as pd

# Load spacy model
nlp = spacy.load('en_core_web_sm')

# Function to remove stopwords from a sentence
def process_text(text):
    doc = nlp(text)
    filtered_words = [token.text for token in doc if not token.is_stop]
    return ' '.join(filtered_words)

if __name__ == "__main__":
    # Create or read the DataFrame 'df' with the 'Review' column
    # Example: df = pd.DataFrame({'Review': ['I went to the stores', 'He visited the parks', 'They ate pizzas', 'It is very beautiful and wonderfully done.']})

    # Apply the process_text function to the 'Review' column
    df['Review'] = df['Review'].apply(lambda x: process_text(str(x)))

    print(df)
```

```python
# exclude the word not
import spacy
import pandas as pd

# Load spacy model
nlp = spacy.load('en_core_web_sm')

# Function to remove stopwords and the word "not" from a sentence
def process_text(text):
    doc = nlp(text)
    filtered_words = [token.text for token in doc if not token.is_stop and token.text.lower() != 'not']
    return ' '.join(filtered_words)

if __name__ == "__main__":
    # Create or read the DataFrame 'df' with the 'Review' column
    # Example: df = pd.DataFrame({'Review': ['I went to the stores', 'He visited the parks', 'They ate pizzas', 'It is very beautiful and wonderfully done.']})

    # Apply the process_text function to the 'Review' column
    df['Review'] = df['Review'].apply(lambda x: process_text(str(x)))

(df)
```

# **Removing unwanted **

```python
# remove empty rows after preprocessing
df.dropna(subset=['Review'], inplace=True)
df = df[~(df['Review'] == '[]')]
```

```python
df['Review'] = df['Review'].str.strip('[]')
```

```python
# remove anyword with less than 3 characters
def remove_short_words(text):
    if text is None:
        return ''
    words = text.split()
    words = [word for word in words if len(word) > 2]
    return ' '.join(words)
```

```python
df['Review'] = df['Review'].apply(remove_short_words)
```
